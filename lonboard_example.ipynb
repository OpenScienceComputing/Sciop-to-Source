```python
import lonboard
import httpx
```

```python
titiler_endpoint = "https://titiler.usgs.gov"
#titiler_endpoint = "https://titiler.xyz"
cog = 'https://data.source.coop/alexgleith/tasmania-dem-2m/Tasmania_Statewide_2m_DEM_14-08-2021.tif'
tilemap = f"{titiler_endpoint}/cog/tiles/{{z}}/{{x}}/{{y}}?url={cog}"
```

```python
viewer = f"{titiler_endpoint}/cog/viewer?url={cog}&algorithm=hillshade"
viewer
```

```python
tilemap
```

```python
tilemap = f'{tilemap}&rescale=0,1600'
tilemap
```

```python
info = f"{titiler_endpoint}/cog/info?url={cog}"
info
```

```python
r = httpx.get(info, verify=False).json()
```

```python
bounds = r["bounds"]
print(bounds)
```

```python
lon_mid = 0.5 * (bounds[0] + bounds[2])
lat_mid = 0.5 * (bounds[1] + bounds[3])
```

```python
view_state={
    "longitude": lon_mid,   
    "latitude": lat_mid,
    "zoom": 3,
#    "pitch": 50.6,
#    "bearing": 8.24,
}
```

```python
from lonboard import Map, BitmapTileLayer

# We set `max_requests < 0` because `tile.openstreetmap.org` supports HTTP/2.
layer = BitmapTileLayer(
    data=tilemap,
    tile_size=256,
    max_requests=-1,
    min_zoom=0,
    max_zoom=19,
    extent=bounds
)

#m = Map(layer)
m = Map(layer, view_state=view_state)
```
